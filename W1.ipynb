```python
# Q6

result = df.loc[(df['Element'] == 'Protein supply quantity (g/capita/day)') & (df['Area'] == 'Madagascar'), 'Y2015'].sum()

# Print the result
print('Total Protein supply quantity in Madagascar in 2015:',result)
```

```python
#Q7

gp=df.groupby(['Y2017','Area'])['Area'].sum()
gp
```

```python
sc = df[['Y2017', 'Area']]
sum_a = sc.groupby('Area').sum()
h_sum = sum_a.idxmax()[0]
print("Area with the highest sum in 2017:", h_sum)
```

```python
#Q8
df.describe()
```

```python
mean_2017 = df['Y2017'].mean()
std_dev_2017 = df['Y2017'].std()

mean_2017_rounded = round(mean_2017, 2)
std_dev_2017_rounded = round(std_dev_2017, 2)

print("Mean for the year 2017:", mean_2017_rounded)
print("Standard deviation for the year 2017:", std_dev_2017_rounded)
```

```python

#Q10
df1 = df[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018', 'Element Code']]

correlation_with_column1 = df1.corr()
correlation_with_column1
```

```python
#Q12

wine=df.groupby('Item')[['Y2015','Y2018']].sum()
print("Total sum of Wine produced in 2015 and 2018 respectively:", wine)
```

```python
#Q14

sel_c = df[['Y2017', 'Area']]

sum_2017 = sel_c.groupby('Area').sum()

low_in = sum_2017['Y2017'].nsmallest(7).index[-1]

print("the 7th lowest sum in 2017:", low_in)
```

```python
#Q17

# Filter the DataFrame for rows where 'Element' is 'Processing'
processing_2017 = df[(df['Element'] == 'Processing') & (df['Y2017'] == 2017)]

wine=processing_2017.groupby('Element').sum()
wine
```

```python
#Q18
missing_2014 = df['Y2014'].isna().sum()

total_rows = len(df)

percentage_missing_2014 = (missing_2014 / total_rows) * 100

print("Total number of missing data in 2014:", missing_2014)
print("Percentage of missing data in 2014: {:.3f}%".format(percentage_missing_2014))
```
